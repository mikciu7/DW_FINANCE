```python
import matplotlib.pyplot as plt

def plot_values_over_time(qp, start_year, start_quarter, end_year, end_quarter, plot_value):
    years_qtrs = []
    values = []

    year, quarter = start_year, start_quarter
    while (year < end_year) or (year == end_year and quarter <= end_quarter):
        label = f"{year}Q{quarter}"
        print(f"Pobieram {label}...")
        try:
            df = qp.get_quarterly_report("MSFT", year, quarter)
            if df is not None and "label" in df.columns:
                row = df.loc[df["label"].isin(plot_value)]
                if not row.empty:
                    value = row.iloc[0].tolist()[-1]
                    years_qtrs.append(label)
                    values.append(value)
        except Exception as e:
            print(f"Błąd {label}: {e}")

        # przejście do następnego kwartału
        quarter += 1
        if quarter > 4:
            quarter = 1
            year += 1

    if not values:
        print("Brak danych do narysowania.")
        return

    plt.figure(figsize=(12, 5))
    plt.plot(years_qtrs, values, marker="o")
    plt.title(f"Apple – {plot_value[0] if len(plot_value) > 1 else plot_value} (kolejne kwartały)")
    plt.xlabel("Kwartał")
    plt.ylabel("Net Income")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_values_over_time(qp, 2009, 2, 2025, 2, ["Revenue", "Contract Revenue"])
```
